```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, Flatten
```

```python
(X_train, Y_train), (X_test, Y_test) =  keras.datasets.mnist.load_data()
```

```python
X_train[0]
```

```python
import matplotlib.pyplot as plt
plt.imshow(X_train[2])
plt.show()
```

```python
X_train = X_train/255.0  # Normalize the data
X_test = X_test/255.0  # Normalize the data
```

```python
model = Sequential()
model.add(Flatten(input_shape=(28, 28)))  # Flatten the input
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))  # Hidden layer with 64 neurons
model.add(Dense(32, activation='relu'))  # Hidden layer with 32 neurons
model.add(Dense(10,activation='softmax'))  # Output layer for 10 classes
```

```python
model.summary()
```

```python
# model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])  # Compile the model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])  # Compile the model
```

```python
his = model.fit(X_train, Y_train, epochs=20, validation_split=0.2, batch_size=128)  # Train the model
```

```python
y_prod = model.predict(X_test)
```

```python
y_pred = y_prod.argmax(axis=1)  # Get the predicted class labels
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_test, y_pred)
accuracy
```

```python
plt.plot(his.history['accuracy'], label='train accuracy')
plt.plot(his.history['val_accuracy'], label='validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
plt.plot(his.history['loss'], label='train loss')
plt.plot(his.history['val_loss'], label='validation loss')
plt.xlabel('Epochs')
```
